```python
import pandas as pd

df = pd.read_csv("taco-db-nutrientes.csv", sep=",", encoding="utf-8", on_bad_lines="skip")
display(df.columns)
# printei as colunas e vi o espaço em braco

# trasformei para string para tirar os espaços em branco com strip
colunas = df.columns.str.strip(' ')
df.columns = colunas
display(df.columns)
```

```python
#removi as indesejadas
removidas = ['Umidade (%)', 'Energia (kcal)', 'Energia (kJ)','Cinzas (g)','Magnésio (mg)', 'Unnamed: 13','Colesterol (mg)']
for coluna in removidas:
    df = df.drop(coluna, axis=1)
display(df.head())
```

```python
for coluna in df.columns:
    print(f"Valores mais frequentes em '{coluna}':")
    print(df[coluna].value_counts().head())  # Exibe os 5 valores mais frequentes
    print("-" * 40)
```

```python
# Substitui 'Tr', 'NA', etc. por zero (ou np.nan se quiser tratar depois)
colunas_nutricionais = [
    "Proteína (g)",
    "Lipídeos (g)",
    "Cálcio (mg)",
    "Carboidrato (g)",
    "Fibra Alimentar (g)"
]

df[colunas_nutricionais] = df[colunas_nutricionais].apply(lambda col: col.astype(str).str.strip())
df[colunas_nutricionais] = df[colunas_nutricionais].replace(['Tr', 'NA', 'NaN', 'nan', '','*'], 0)

for col in colunas_nutricionais:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[colunas_nutricionais] = df[colunas_nutricionais].fillna(0)
```

```python

print("Valores nulos por coluna (NaN):")
print(df.isnull().sum())
print("\nNúmero de valores NaN por coluna:")
print(df.isna().sum()) 
print("\nDimensões do DataFrame (linhas, colunas):")
print(df.shape)
print("\nCabeçalho do DataFrame (nomes das colunas):")
print(df.head(0))
```

```python
df.to_parquet("taco-db-nutrientes.parquet", index=False)
```
